```python
import matplotlib.pyplot as plt
import numpy as np

def resoudre_systeme_2x2():
    # Lecture des coefficients
    print("Entrez les coefficients du système d'équations:")
    a1 = float(input("a1: "))
    b1 = float(input("b1: "))
    c1 = float(input("c1: "))
    a2 = float(input("a2: "))
    b2 = float(input("b2: "))
    c2 = float(input("c2: "))
    
    # Calcul du déterminant
    determinant = a1 * b2 - a2 * b1
    
    # Vérification de la solvabilité
    if determinant == 0:
        print("\nLe système n'a pas de solution unique (déterminant nul).")
        print("Les droites sont parallèles ou confondues.")
        
        # Visualisation même si pas de solution unique
        plt.figure(figsize=(8, 6))
        plt.title("Représentation graphique des deux équations")
        plt.xlabel("x")
        plt.ylabel("y")
        plt.grid(True)
        
        x_vals = np.linspace(-10, 10, 100)
        
        if b1 != 0:
            y1 = (c1 - a1 * x_vals) / b1
            plt.plot(x_vals, y1, label=f'{a1}x + {b1}y = {c1}')
        else:
            plt.axvline(x=c1/a1, label=f'{a1}x = {c1}')
        
        # Équation 2
        if b2 != 0:
            y2 = (c2 - a2 * x_vals) / b2
            plt.plot(x_vals, y2, label=f'{a2}x + {b2}y = {c2}')
        else:
            plt.axvline(x=c2/a2, label=f'{a2}x = {c2}')
        
        plt.legend()
        plt.show()
        return 
        
    x = (b2 * c1 - b1 * c2) / determinant
    y = (a1 * c2 - a2 * c1) / determinant
    
    # Affichage des résultats
    print(f"\nSolution du système:")
    print(f"x = {x:.2f}")
    print(f"y = {y:.2f}")
    
    plt.figure(figsize=(8, 6))
    plt.title("Représentation graphique des deux équations")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    
    x_vals = np.linspace(x-5, x+5, 100)
    
    if b1 != 0:
        y1 = (c1 - a1 * x_vals) / b1
        plt.plot(x_vals, y1, label=f'{a1}x + {b1}y = {c1}')
    else:
        plt.axvline(x=c1/a1, label=f'{a1}x = {c1}')
        
    if b2 != 0:
        y2 = (c2 - a2 * x_vals) / b2
        plt.plot(x_vals, y2, label=f'{a2}x + {b2}y = {c2}')
    else:
        plt.axvline(x=c2/a2, label=f'{a2}x = {c2}')
    
    plt.plot(x, y, 'ro', label=f'Solution ({x:.2f}, {y:.2f})')
    
    plt.legend()
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.show()
resoudre_systeme_2x2()
```
